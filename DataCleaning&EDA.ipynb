## Importing the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

## Data cleaning

```python
## Missing values in any column

df[df.isnull().any(axis=1)]
```

## The dataset is having two dataset partitions, Bejaia regin and Sidi-Bel region

The dataset will be converted now into 2 sets bassed on region from 122 index, we can make a new column based on the region

1. "Bejaia Region Dataset".
2. "Sidi-Bel Abbes Region Dataset"

## Adding new column with region. i.e; "Bejaia Region Dataset" ----> 0 value and "Sidi-Bel Abbes Region Dataset" --> 1 value

```python
### As till 122 index it is Bejaia and after that it is Sidi-bel. So we will take the index number for creating the region column

df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df.info()
```

```python
## Converting the "Region" field into integer
df['Region'] = df['Region'].astype(int)
df.info()
```

```python
df.isnull().sum()
```

```python
### Droping the null values
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
### Removing the 122 index row
df=df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
### Removing extra spaces in column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

### changing the required columns as integer data type

```python
df.head()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### changing the remaining columns to float datatype

```python
newfloat=[]
for features in df.columns:
    if df[features].dtype == 'O':
        newfloat.append(features)
```

```python
for i in newfloat:
    if i!= 'Classes':
        df[i]=df[i].astype(float)
```

```python
### Monthly fire analysis
#---------------------we will use df, as we have monthly data there, in df_copy we have removed the day,month,year
dftemp=df.loc[df['Region']==0]
df['Classes']=np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
sns.set_style('darkgrid')
sns.countplot(data=dftemp, x='month', hue='Classes')
plt.title("Fire amalysis for Bejaia region")
plt.ylabel('Number of fire', weight='bold')
plt.xlabel('Month', weight='bold')
```

From the analysis we can see that in the Month of August the fire caught the forest of Bejaia region the most
